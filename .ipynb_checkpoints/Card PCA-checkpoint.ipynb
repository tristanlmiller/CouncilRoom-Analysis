I want to apply PCA to the cards.  Specifically, I would like to use each row in the synergy-factor matrix as vectors for each card.  My initial guess is that the first component will be "villaginess vs terminalness" and the second component will be "slogginess vs thinness".

I also need to visually represent the results.  Perhaps a labeled graph would work but maybe there would be too many words.  Could also just print out a table of coordinates, or else I could make word clouds.

Another possibility is to show synergies between card types.  Do terminals synergize with villages?

```python
#packages
import os
import re
import pdb
import numpy as np
import pickle
import time
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import sklearn.semi_supervised as semi
from sklearn.cluster import SpectralClustering as spectral
```

## Preparing data

```python
#load pickles
with open('gain_matrices.pkl','rb') as f:
    num_games, game_gains, total_gains = pickle.load(f)
with open('card_list.pkl','rb') as f:
    card_list,card_dict = pickle.load(f)
    
#Exclude contribution from non-kingdom cards
card_excluder = np.ones((len(card_list)))
excluded_cards = ['Copper','Silver','Gold','Estate','Duchy','Province','Curse','Colony','Platinum','Ruins','Potion']
#excluded_cards = ['Copper','Silver','Gold','Estate','Duchy','Province','Curse','Colony','Platinum','Ruins','Potion','Prince','Walled Village']
#We can keep the low-statistics cards, we'll just weight everything by their statistics

for card in excluded_cards:
    card_excluder[card_dict[card]] = 0
    
#element-wise division
prc_gains = game_gains / num_games

#Exclude non-kingdom cards by using card_excluder
synergies = prc_gains * card_excluder

#Now from each row, subtract the vector from the average game
base_gain = synergies[card_dict['Copper'],:].copy()
synergies -= base_gain

#Normalize each row
syn_factor = np.sum(synergies**2,axis = 1)**0.5
synergies = (synergies.transpose() / syn_factor).transpose()

#weight columns by their statistics
#basically, each vector will lie on the surface of an ellipsoid
stat_weight = np.mean(num_games**0.5,axis=1)
weight_norm = np.mean([stat_weight[i] for i in range(len(stat_weight)) if card_excluder[i] != 0])
stat_weight = stat_weight / weight_norm
synergies = synergies * stat_weight
```

```python
#put data in DataFrame
syn_vectors = pd.DataFrame({card_list[i]:synergies[:,i] for i in range(len(card_list))})
#syn_vectors['Card'] = card_list
#syn_vectors = syn_vectors.set_index('Card')
syn_vectors.index = card_list

#Drop irrelevant rows and columns
non_kingdom_cards = ['Copper','Silver','Gold','Estate','Duchy','Province','Curse','Colony','Platinum','Ruins','Potion']
syn_vectors = syn_vectors.drop(non_kingdom_cards,axis=0)
syn_vectors = syn_vectors.drop(excluded_cards,axis=1)

#recalculate card_list and card_dict for just the kingdom cards.  rename to kingdom_list and kingdom_dict
kingdom_list = syn_vectors.index.tolist()
kingdom_dict = {card:i for i,card in enumerate(kingdom_list)}
```

## Applying PCA

```python
start_time = time.time()
model = PCA(n_components=10)
reduced_vectors = model.fit_transform(syn_vectors)
enabled_cards = model.components_
processing_time = (time.time() - start_time)/60
print("Time to execute: %.2f minutes" % processing_time )
```

```python
enabled_cards.shape
```

```python
%matplotlib inline

#Plot any two components.  The labels I'm using are 1-indexed instead of 0-indexed
#if axis1 or axis2 is negative, it will show the cards most promoted by that component instead
def pca_map(axis1,axis2):
    plt.figure(figsize=(18,13))
    if( axis1 > 0 ):
        component1 = reduced_vectors[:,axis1-1]
        plt.xlabel('Component %i' % axis1,fontsize=16)
    else:
        axis1 = -axis1
        component1 = enabled_cards[axis1-1,:].transpose()
        plt.xlabel('Cards promoted by component %i' % axis1,fontsize=16)
    
    if( axis2 > 0):
        component2 = reduced_vectors[:,axis2-1]
        plt.ylabel('Component %i' % axis2,fontsize=16)
    else:
        axis2 = -axis2
        component2 = enabled_cards[axis2-1,:].transpose()
        plt.ylabel('Cards promoted by component %i' % axis2,fontsize=16)
        
    plt.scatter(component1, component2)
    for label, x, y in zip(kingdom_list, component1, component2):
        #plt.text(x,y+.1,label,ha='center', va='center')
        plt.annotate(label, xy = (x,y),xytext = (0,5),textcoords='offset points',ha='center',fontsize=12)
    plt.show()
```

```python
pca_map(1,2)
pca_map(3,4)
```

# Clustering
Since I have some idea for what kind of clusters I want, it would be nice to come up with some sort of semi-supervised clustering model.

One interesting possibility is label propagation, from the scikit-learn package:
http://scikit-learn.org/stable/modules/label_propagation.html

```python
#seeds for the clusters
#village,term draw, nonterm draw, trash, junk, gain,term payoff, cantrip payoff, stop card payoff, victory
cluster_seeds = [['Village','Native Village','Throne Room'],['Smithy','Courtyard','Catacombs'],
                 ['Laboratory','Caravan','Stables'],['Chapel','Moneylender','Lookout'],
                ['Witch','Sea Hag','Mountebank'],['Workshop','Ironworks','Smugglers'],
                ['Bridge','Butcher','Bank'],['Gardens','Duke','Fairgrounds']]

#cluster_seeds = ['Village','Smithy','Laboratory','Chapel','Witch','Workshop','Woodcutter','Market','Gardens']

labels = np.zeros((len(kingdom_list)),dtype=int)-1
for i,seed in enumerate(cluster_seeds):
    #labels[kingdom_dict[seed]] = i
    for card in seed:
        labels[kingdom_dict[card]] = i
```

```python
label_model = semi.LabelPropagation()
label_model.fit(syn_vectors,labels)
```

```python
#Get a list of cards in each cluster
clusters = [[] for i in range(len(cluster_seeds))]
for i,cluster_id in enumerate(label_model.transduction_):
    clusters[cluster_id].append(kingdom_list[i])
```

```python
#print out the list
output_string = ''
for i,cluster in enumerate(clusters):
    output_string += '%s cluster:\n' % cluster_seeds[i]
    for card in cluster:
        output_string += card + '\t'
    output_string += '\n\n'
print(output_string)
```

## Spectral Clustering
What if I try without any labeling?

```python
num_clusters = 9
spectral_model = spectral(n_clusters=num_clusters)
cluster_labels = spectral_model.fit_predict(reduced_vectors)
```

```python
clusters = [[] for i in range(num_clusters)]
for i,cluster_id in enumerate(cluster_labels):
    clusters[cluster_id].append(kingdom_list[i])
    
output_string = ''
for i,cluster in enumerate(clusters):
    output_string += 'Cluster %i:\n' % (i+1)
    for card in cluster:
        output_string += card + '\t'
    output_string += '\n\n'
print(output_string)
```
